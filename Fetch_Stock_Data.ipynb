```python
#Importing Libraries (Numpy, Pandas & Matplotlib are IMPORTANT)
import numpy as np #linear algebra
import yfinance as yf #fetching stock data
from yfinance import ticker
import pandas as pd #data manipulation
import seaborn as sns #data visualisation
import matplotlib.pyplot as plt #data visualisation from Matlab
```

```python
#Let's define Stock Data
def stock_data(ticker, start_date, end_date):
    """"
    Let's fetch historical stock data from Yahoo Finance & visualise it."

    Important Parameters:
    (")
    ticker (str): Stock Ticker symbol (e.g. 'AAPL' for Apple Inc.)
    start_date (str): Starting date (e.g. '2022-01-01')
    end_date (str): End date (e.g. '2022-12-31')
    Never forget:
    1) start_date is inclusive and end_date is exclusive
    2) The stock market trades for 252 days in a year

    Returns:
    DataFrame: Pandas DataFrame containing historical stock data
    """
    print(f"Fetching data for {ticker} from {start_date} to {end_date}")
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    print(f"Data fetched for {ticker} from {start_date} to {end_date}")
    return stock_data
```

```python
#Let's Plot the stock data
def plot_stock_data(stock_data, ticker):
    """
    Let's plot the stock data over time
    """
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data['Close'], color='blue', label='Close Price')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(f"{ticker} Close Price")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_average_daily_returns(stock_data, ticker):
    """"
    Let's plot the average daily returns over time
    """
    stock_data['Daily_Returns'] = stock_data['Close'].pct_change()
    avg_daily_returns = stock_data["Daily_Returns"].mean()
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data["Daily_Returns"], color='blue', label='Daily Returns')
    plt.axhline(y=avg_daily_returns, color='red', linestyle='--', label='Average Daily Returns')
    plt.xlabel('Date')
    plt.ylabel('Returns')
    plt.title(f'{ticker} Daily Returns')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
if __name__ == "__main__":
    print("Welcome to Stock Data Analysis")

    #Stock Details
    ticker = 'AAPL'
    start_date = '2022-01-01'
    end_date = '2022-12-31'

    stock_data= stock_data(ticker, start_date, end_date)

    if not stock_data.empty:
        plot_stock_data(stock_data, ticker)
        plot_average_daily_returns(stock_data, ticker)
    else:
        print("No Data Found")
```
